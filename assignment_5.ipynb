### Q1) Write a program that has classes such as Student, Course and Department. Enroll a student in a course of a particular department.

```python
class Department:
    def __init__(self, name):
        self.name = name
        self.courses = []

    def add_course(self, course):
        self.courses.append(course)


class Course:
    def __init__(self, course_name, dept):
        self.course_name = course_name
        self.department = dept
        self.students = []
        dept.add_course(self)

    def enroll_student(self, student):
        self.students.append(student)


class Student:
    def __init__(self, name):
        self.name = name

    def enroll(self, course):
        course.enroll_student(self)
        print(f"{self.name} enrolled in {course.course_name} ({course.department.name})")



dept = Department("Computer Science")
course1 = Course("Python Programming", dept)

s1 = Student("Alice")
s1.enroll(course1)
```

### Q2) Write a program with class Bill. The users have the option to pay the bill either by cheque or by cash. Use the inheritance to model this situation.

```python
class Bill:
    def __init__(self, amount):
        self.amount = amount

    def pay(self):
        pass


class CashPayment(Bill):
    def pay(self):
        print(f"Bill of ₹{self.amount} paid by CASH.")


class ChequePayment(Bill):
    def __init__(self, amount, cheque_no):
        super().__init__(amount)
        self.cheque_no = cheque_no

    def pay(self):
        print(f"Bill of ₹{self.amount} paid by CHEQUE (No: {self.cheque_no}).")



b1 = CashPayment(500)
b1.pay()

b2 = ChequePayment(1200, "CHQ9832")
b2.pay()
```

### Q3) Write a program to overload the - =operator to subtract two Distance Objects.

```python
class Distance:
    def __init__(self, meters):
        self.meters = meters

    def __isub__(self, other):
        self.meters -= other.meters
        return self

    def __str__(self):
        return f"{self.meters} meters"



d1 = Distance(150)
d2 = Distance(40)

d1 -= d2
print("After subtraction:", d1)
```
