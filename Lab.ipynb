```python
# WAP to sum series
s1=s2=0
for i in range(1,101):
    s1+=i/(i+1)
print(" Sum of series 1 : ", s1)

for i in range(1,101):
    s2+=(i*i)/i
print(" Sum of series 2 : ", s2)

```

```python
# WAP to print a 3x3 grid

for i in range(3):
    for j in range(3):
        print("(", i, " ,", j , ")" , end=" ")
    print()
```

```python
# WAP to perform matrix multiplication

a = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

b = [
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1]
]

result = [[0 for _ in range(len(b))] for _ in range(len(a))]

for i in range(len(a)):
    for j in range(len(b)):
        for k in range(len(a)):
            result[i][j] += a[i][k] *  b[k][j]

result
```

```python
# WAP to print star pattern

for i in range(6):
    for j in range(i):
        print(" *", end=" ")
    print()
for i in range(6):
    for j in range(6 - i):
        print(" *", end=" ")
    print()
```

```python
# WAP Genrating Combination

adjectives = ["red", "big", "tasty"]
fruits = ["apple", "bannana", "cherry"]

for adj in adjectives:
    for fruit in fruits:
        print(adj,fruit)
```

```python
# Wap dual piramid

for i in range(6):
    print(" "*(6-i), end="")
    for j in range(i):
        print("* ", end="")
    print()

for i in range(6):
    print(" "*(i), end="")
    for j in range(6-i):
        print("* ", end="")
    print()
```

```python
# WAP fibbonaci
a, b= 0, 1
for i in range(0,10):
    print(a, end=" ")
    a, b = b, a+b
```

```python
# WAP to print prime numbers
for i in range(2, 51):
    for j in range(2, i // 2 + 1):
        if i % j == 0:
            break
    else:
        print(i, end=" ")
```

```python
# Pattern

for i in range(3):
    for j in range(3):
        if(i==j):
            print("*", end=" ")
        else:
            print("-", end=" ")
    print()
```

```python
# WAP to print hollow square

for i in range(5):
    for j in range(5):
        if(i==0 or j==0 or i == 4 or j==4):
            print("*", end=" ")
        else:
            print(" ", end=" ")
    print()
```
