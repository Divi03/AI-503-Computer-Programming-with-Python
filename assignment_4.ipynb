### Q1) Write a Python function find_smallest_multiple(n: int) that uses a while loop to find the smallest
positive integer x such that x is divisible by all numbers from 1 to n.
Conditions:
- i. The function should keep incrementing x by 1 until it finds a number that satisfies the condition.
- ii. Your solution should handle the edge case where n = 1 efficiently, returning 1 directly since 1 is divisible by itself.

```python
def find_smallest_multiple(n: int) -> int:
    if n == 1:
        return 1
    x = 1
    while True:
        divisible = True
        for i in range(1, n + 1):
            if x % i != 0:
                divisible = False
                break
        if divisible:
            return x
        x += 1

# Example:
print(find_smallest_multiple(5))  # Output: 60
```

### Q2) Write a function cubesum() that accepts an integer and returns the sum of the cubes of individual digits
of that number. Use this function to make functions PrintArmstrong() and isArmstrong() to print
Armstrong numbers and to find whether is an Armstrong number.

```python
def cubesum(num: int) -> int:
    return sum(int(digit)**3 for digit in str(num))

def isArmstrong(num: int) -> bool:
    return cubesum(num) == num

def PrintArmstrong(limit: int):
    for i in range(limit + 1):
        if isArmstrong(i):
            print(i, end=" ")
    print()

# Example:
PrintArmstrong(500)  # Prints all Armstrong numbers up to 500
```

### Q3) Why is operator precedence important? Give an example where neglecting precedence changes the result.

```python
"""
Operator precedence determines the order in which expressions are evaluated. Neglecting it can lead to wrong results.
"""

result1 = 2 + 3 * 4  # Multiplication first → 2 + 12 = 14
result2 = (2 + 3) * 4  # Parentheses first → 5 * 4 = 20

print(result1, result2)  # 14 20
```

### Q6) Write a program that takes a sentence as input and counts the frequency of vowels, consonants, digits,
and special characters separately

```python
sentence = input("Enter a sentence: ")
vowels = consonants = digits = special = 0

for char in sentence:
    if char.isdigit():
        digits += 1
    elif char.isalpha():
        if char.lower() in 'aeiou':
            vowels += 1
        else:
            consonants += 1
    else:
        special += 1

print("Vowels:", vowels)
print("Consonants:", consonants)
print("Digits:", digits)
print("Special characters:", special)
```

### Q7) Write a Python program to create a dictionary of students’ names as keys and their marks as values.
Then:
- A. Print the student with the highest marks
- B. Print the student with the lowest marks

```python
students = {
    "Alice": 85,
    "Bob": 92,
    "Charlie": 78,
    "David": 95
}

# A. Student with highest marks
highest = max(students, key=students.get)
print("Highest marks:", highest, students[highest])

# B. Student with lowest marks
lowest = min(students, key=students.get)
print("Lowest marks:", lowest, students[lowest])
```
