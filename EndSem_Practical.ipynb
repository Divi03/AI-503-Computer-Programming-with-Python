### Q1 Write a Python program to find the sum of the digits of factorial of a given number.


```python
def factorial(n):
    """
    Calculate the factorial of a number n.
    factorial(n) = 1 * 2 * 3 * ... * n
    """
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result

def sum_of_digits_of_factorial(n):
    """
    Calculate the sum of the digits of n! using simple loops.
    """
    fact = factorial(n)
    total = 0
    for digit in str(fact):
        total += int(digit)

    # Step 3: print the result
    print(f"The sum of the digits of {n}! is: {total}")


sum_of_digits_of_factorial(5)
```

### Q2 Use bitwise operators to check if a given number is a power of 2

```python
def is_power_of_two(n):
    """
    Check if n is a power of 2.
    A number is a power of two if it has exactly one '1' bit in binary representation.

            8      → 1000
            8 - 1  → 0111
            AND    → 0000
    """
    return n > 0 and (n & (n - 1)) == 0

number = 64
if is_power_of_two(number):
    print(f"{number} is a power of two.")
else:
    print(f"{number} is NOT a power of two.")
```

### Q3 Write a Python program to find the sum of all the positive integers which cannot be written as the sum of two abundant numbers.

```python
"""
This function calculates the sum of proper divisors of a given number.
Proper divisors include all positive divisors except the number itself.
The function checks divisors only up to the square root of the number
to improve efficiency.
"""
def sum_of_divisors(n):
    total = 1
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            total += i
            if i != n // i:
                total += n // i
    return total


"""
This function generates a list of all abundant numbers up to a given limit.
A number is considered abundant if the sum of its proper divisors
is greater than the number itself.
The smallest abundant number is 12, so the search starts from there.
"""
def get_abundant_numbers(limit):
    abundant_nums = []
    for i in range(12, limit + 1):
        if sum_of_divisors(i) > i:
            abundant_nums.append(i)
    return abundant_nums


"""
This function calculates the sum of all positive integers up to the given limit
that cannot be written as the sum of two abundant numbers.
It first finds all abundant numbers, then marks all possible sums of
two abundant numbers, and finally adds the numbers that remain unmarked.
"""
def non_abundant_sums(limit):
    abundant_nums = get_abundant_numbers(limit)
    can_be_written = [False] * (limit + 1)

    for i in range(len(abundant_nums)):
        for j in range(i, len(abundant_nums)):
            s = abundant_nums[i] + abundant_nums[j]
            if s <= limit:
                can_be_written[s] = True

    total = 0
    for i in range(1, limit + 1):
        if not can_be_written[i]:
            total += i

    return total


# Given upper limit (mathematically proven bound)
limit = 28123

# Display the final result
result = non_abundant_sums(limit)
print("The sum of all positive integers which cannot be written as the sum of two abundant numbers is:", result)
```
