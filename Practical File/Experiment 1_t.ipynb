```python
# Displaying different data types using variables and literals

integer_var = 25
float_var = 12.5
string_var = "Python Programming"
boolean_var = True

print("Integer value:", integer_var)
print("Float value:", float_var)
print("String value:", string_var)
print("Boolean value:", boolean_var)

# Literal constants
print("Integer literal:", 100)
print("Float literal:", 3.14)
print("String literal:", "Hello World")
print("Boolean literal:", False)

"""
This program demonstrates how Python supports different data types such as integer, 
float, string, and boolean. Variables are created and assigned values directly without 
specifying data types because Python is dynamically typed. The print() function outputs 
all values, showing how Python can handle multiple data types simultaneously.

"""
```

```python
# Using += operator on strings

str1 = "Hello"
str1 += " World"

print("Updated string:", str1)

"""
The += operator is used here for string concatenation. Initially, 
a string variable stores a value. Using += appends another string to the existing 
value and updates the same variable. This shows that strings can be combined using 
assignment operators.

"""
```

```python
# String addition and multiplication

str1 = "Python"
str2 = " Lab"

# Addition (Concatenation)
add_result = str1 + str2

# Multiplication (Repetition)
mul_result = str1 * 3

print("String Addition Result:", add_result)
print("String Multiplication Result:", mul_result)


"""
String addition performs concatenation, joining two strings into one. 
String multiplication repeats the string a specified number of times. 
This program demonstrates both behaviors, highlighting Pythonâ€™s operator overloading 
for string objects.

"""
```
