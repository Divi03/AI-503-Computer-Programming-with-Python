```python
# Calculate area of a triangle using Heron's Formula

a = float(input("Enter first side: "))
b = float(input("Enter second side: "))
c = float(input("Enter third side: "))

# Semi-perimeter
s = (a + b + c) / 2

# Area calculation
area = (s * (s - a) * (s - b) * (s - c)) ** 0.5

print("Area of the triangle is:", area)

"""
The program calculates the area of a triangle when all three sides are known. 
First, it computes the semi-perimeter using Heronâ€™s formula. 
Then, it applies the square root formula using the math module. 
This demonstrates arithmetic operators, user input, and mathematical functions.

"""
```

```python
# Calculate total money in a piggy bank

rs10 = int(input("Enter number of Rs.10 coins: "))
rs5 = int(input("Enter number of Rs.5 coins: "))
rs2 = int(input("Enter number of Rs.2 coins: "))
rs1 = int(input("Enter number of Rs.1 coins: "))

total_amount = (rs10 * 10) + (rs5 * 5) + (rs2 * 2) + (rs1 * 1)

print("Total amount in piggy bank is: Rs.", total_amount)

"""
This program calculates the total money based on the number of coins of different denominations.
Each coin count is multiplied by its value and added together. 
It demonstrates arithmetic operations and real-life financial calculations.

"""
```
